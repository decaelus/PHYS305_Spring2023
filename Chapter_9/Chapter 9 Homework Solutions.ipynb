### 9.2

We're given a magnitude difference between two stars, but all of the thinking we have in this chapter involves fluxes. So the first thing we need to do is convert the magnitude difference $\Delta m$ into a relationship in terms of (incorrect) fluxes, $F_1^\prime$ and $F_2^\prime$ -- "incorrect" because they are impacted by the uncorrected non-linear response.

$$
\Delta m^\prime = -2.5 \log_{10} \left( \frac{F_1^\prime}{F_2^\prime} \right) \Rightarrow \frac{F_2^\prime}{F_1^\prime} = 10^{\Delta m/2.5} = 10^{1.25/2.5} = 3,
$$
meaning star 2 is *measured* to be 3 times brighter than star 1, but because of the non-linearity of our detector, the apparent flux for star 2 is incorrectly estimated.

We're told that star 2 produced a DN value of 30000, meaning star 1 produced a DN value of 10000. 

If $F_{1/2}$ is the *corrected* relative flux for star 1/2, then the corresponding output, i.e., data number ${\rm DN}_{1/2}$ is given by Equation 9.23:
$$
{\rm DN}_{1/2} = a + b \left( F_{1/2} t \right) + c \left( F_{1/2} t\right)^2,
$$
where $t$ is the exposure time (and we're assuming exposure times for both stars are the same, as indicated).

We can use the quadratic equation to calculate $F_{1/2} t$:
$$
F_{1/2}t = \frac{-b \pm \sqrt{b^2 - 4 \left( a - {\rm DN}_{1/2} \right) c}}{ 2 c }.
$$

For star 2, we have
$$
F_2 t = \frac{-533 \pm \sqrt{\left( 533 \right)^2 - 4 \left( -39.6 - 30000 \right) \left( -1.54 \right)}}{2 \left( -1.54 \right)} = 275\ {\rm or}\ 70.9.
$$

For star 1,
$$
F_1 t = \frac{-533 \pm \sqrt{\left( 533 \right)^2 - 4 \left( -39.6 - 10000 \right) \left( -1.54 \right)}}{2 \left( -1.54 \right)} = 326\ {\rm or}\ 20.1.
$$

Using the flux values that are closest to one another means we are assuming (reasonably) that the non-linear effects are small. So
$$
\frac{F_2}{F_1} = \frac{70.9}{20.1} \approx 3.5.
$$

In any case, we can now estimate the correct relative magnitudes:
$$
\Delta m = -2.5 \log_{10} \left( \frac{F_1}{F_2} \right) = -2.5 \log_{10} \left( 1/3.5 \right) \approx \boxed{ 1.36 }.
$$
